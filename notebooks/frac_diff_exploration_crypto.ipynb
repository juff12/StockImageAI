```python
def get_frac_data(data):
    data['fracChange'] = (data['close'].sub(data['open'])).div(data['open'])
    data['fracHigh'] = (data['high'].sub(data['open'])).div(data['open'])
    data['fracLow'] = (data['open'].sub(data['low'])).div(data['open'])
    
    max = data['fracChange'].max()
    min = data['fracChange'].min()
    mean = data['fracChange'].apply(lambda x: abs(x)).mean()
    std = data['fracChange'].std()

    return [max, min, mean, std]
```

```python
def get_fraclow_data(data):
    data['fracChange'] = (data['close'].sub(data['open'])).div(data['open'])
    data['fracHigh'] = (data['high'].sub(data['open'])).div(data['open'])
    data['fracLow'] = (data['open'].sub(data['low'])).div(data['open'])
    
    max = data['fracLow'].max()
    min = data['fracLow'].min()
    mean = data['fracLow'].apply(lambda x: abs(x)).mean()
    std = data['fracLow'].std()

    return [max, min, mean, std]
```

```python
def get_frachigh_data(data):
    data['fracChange'] = (data['close'].sub(data['open'])).div(data['open'])
    data['fracHigh'] = (data['high'].sub(data['open'])).div(data['open'])
    data['fracLow'] = (data['open'].sub(data['low'])).div(data['open'])
    
    max = data['fracHigh'].max()
    min = data['fracHigh'].min()
    mean = data['fracHigh'].apply(lambda x: abs(x)).mean()
    std = data['fracHigh'].std()

    return [max, min, mean, std]
```

```python
tickers = pickle.load(open('data/crypto/iterables/coinbase_tickers.pkl', 'rb'))
```

```python
tickers = pickle.load(open('data/stock/iterables/sp500_tickers.pkl', 'rb'))
```

```python
def run_all_for_time(bartime,parentdir):
    frac_summary_change,frac_summary_high,frac_summary_low = [],[],[]
    for ticker in tickers:
        data = pd.read_csv('data/{p}/data/formatted/{t}/{t}_{b}_data_formatted.csv'.format(p=parentdir,t=ticker,b=bartime))
        frac_summary_change.append(get_frac_data(data))
        frac_summary_high.append(get_frachigh_data(data))
        frac_summary_low.append(get_fraclow_data(data))
        
    return pd.DataFrame(frac_summary_change,index=tickers,columns=['max','min','mean','std']), pd.DataFrame(frac_summary_high,index=tickers,columns=['max','min','mean','std']), pd.DataFrame(frac_summary_low,index=tickers,columns=['max','min','mean','std'])
```
