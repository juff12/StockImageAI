```python
from polygon import RESTClient
import datetime
import pandas as pd
```

```python
client = RESTClient(api_key="")
```

```python
ticker = 'AAPL'
multiplier = 1
timespan = 'day'
start = '2023-01-01'
end = '2024-01-01'
limit = 50000
```

```python
aggs = []
for a in client.list_aggs(ticker=ticker, multiplier=multiplier, timespan=timespan, from_=start, to=end, limit=limit):
    aggs.append(a)
```

```python
def convert_date(mili):
    s = mili / 1000.0
    return datetime.datetime.fromtimestamp(s).strftime('%Y-%m-%d')
```

```python
df = pd.DataFrame(aggs)
df
```

```python
df.to_csv('aapl_data_raw.csv')
```
